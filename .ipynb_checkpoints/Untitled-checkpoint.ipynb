Get stock market data.

```python
import datetime as dt
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
import pytz
yf.pdr_override()
#import pandas.io.data as web
```

# Specify date range for analysis


```python
end = dt.datetime.now()

#start = end - dt.timedelta(days=5000) #5000 days ago from today
start = dt.datetime(2000,1,1)

start, end
```

# Select the Stocks/tikcers

```python
stocklist = ['ASML.AS','ASM.AS']
SP500 = ['^GSPC']
```

# Call the Pandas_DataReader module:


```python
df = pdr.get_data_yahoo(stocklist, start, end)
df2 = pdr.get_data_yahoo(SP500, start, end)
df.head()
```

# Looking at df structure

# Access useful attributes

WE would like to use close price information for all of these stocks, graph these price tracers.

```python
close = df.Close
close.head()
```

# Gain insight quickly with pandas describe method

```python
close.describe()
```

e.g. last 100 days

```python
condition = end - dt.timedelta(days=142)
timezone = pytz.timezone('Europe/Amsterdam') 
condition_aware = timezone.localize(condition)
close[close.index > condition_aware].describe()
```
