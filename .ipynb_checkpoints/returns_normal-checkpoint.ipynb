```python
import datetime as dt
from datetime import datetime
import pandas as pd
import numpy as np
from pandas_datareader import data as pdr
import plotly.offline as pyo
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
pyo.init_notebook_mode(connected=True)
pd.options.plotting.backend = 'plotly'
import yfinance as yf
import pytz
from scipy import stats
yf.pdr_override()
import pylab
```

```python
start = dt.datetime(2000,1,1)
end = dt.datetime.now()
```

```python
stocks = ['^GSPC','TSLA', 'AAPL','AMZN','GOOGL']
stocks.sort()
df = pdr.get_data_yahoo(stocks, start, end)
df.head()
```

```python
def simple_return_dict(stocks=stocks):
    simple_return_dict = {}
    for s in stocks:
        simple_return_dict[s] = ((df.Close[s]-df.Close[s].shift(1))/df.Close[s].shift(1)).dropna()
    return simple_return_dict

def log_return_dict(stocks=stocks):
    log_return_dict = {}
    for s in stocks:
        log_return_dict[s] = (np.log(df.Close[s]/df.Close[s].shift(1))).dropna()
    return log_return_dict




```

# LOG vs SIMPLE RETURNS


PLot financial data and look at first and last share prices. 

```python
df.Close['AAPL'].plot().update_layout(autosize = False, width = 500, height = 300)
```

```python
print('First', df.Close['AAPL'][0], 'Last', df.Close['AAPL'][-1])
```

Use simple returns and attempt to compute final price form starting price over time horizon.

```python
ret_AAPL = simple_return_dict(['AAPL'])
ret_AAPL = ret_AAPL['AAPL']
mean_ret = ret_AAPL.mean()
df.Close['AAPL'][0] * (1+mean_ret)**len(ret_AAPL)
```

This doesnt make much sense, how to actually compute?

```python
df.Close['AAPL'][0] * np.prod([(1 + Rt) for Rt in ret_AAPL])
```

So simple return mean is not a very useful and descriptive,
What about log?

```python
log_ret_AAPL = log_return_dict(['AAPL'])
log_ret_AAPL = log_ret_AAPL['AAPL']
mean_log_ret = log_ret_AAPL.mean()
mean_log_ret
```

```python
df.Close['AAPL'][0]*np.exp(mean_log_ret*len(log_ret_AAPL))
```

# THE MAIN REASON
### If we want to model returns using the normal distribution!
- SIMPLE RETURNS: product of normally distributed variables is NOT normal.
- LOG RETURNS: The sum of normally distributed variables follows a normal d.


Also: log distribution bounds our stock price at 0.

### Histogram of log returns

```python
log_ret_AAPL.plot(kind='hist').update_xaxes(range=[-0.15, 0.15]).update_layout(autosize = False, width = 650, height = 400)
```

Is financial data normal assumption good?
the assum that prices or more accuretly log returns are normally distributed!

```python
log_ret_AAPL_sorted = log_ret_AAPL.tolist()
log_ret_AAPL_sorted.sort()
worst = log_ret_AAPL_sorted[1]
best = log_ret_AAPL_sorted[-1]

std_worst = (worst - mean_log_ret)/log_ret_AAPL.std()
std_best = (best - mean_log_ret)/log_ret_AAPL.std()

print('Std dev. worst %.2f best %.2f'%(std_worst, std_best))
```
